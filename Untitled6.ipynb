```python
# Calcular los valores máximos y mínimos
max_value = df_total['Price'].max()
min_value = df_total['Price'].min()

# Obtener las fechas correspondientes a los valores máximos y mínimos ,
#.values[0] es que cuando seleccionas las fechas correspondientes a los valores máximos y mínimos, obtienes una serie de Pandas, y usar
#values[0] extrae el valor real de esa serie.
max_date = df_total.loc[df_total['Price'] == max_value, 'date'].values[0]
min_date = df_total.loc[df_total['Price'] == min_value, 'date'].values[0]

# Filtrar el DataFrame para obtener solo las filas con fechas máximas y mínimas
df_filtered = df_total.loc[(df_total['date'] == max_date) | (df_total['date'] == min_date)]


df_filtered
```
